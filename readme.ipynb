# About the word Monad

When I think of a monad I think about an apparently random  set of events in the toy-universe that leave as a trace of it's occurrence a set of indivisible points that have some information associated to them, and this information can be thought as a biased-reflection of it's immediate context and the whole system state, but at the same time has it's own essence given by the "architect" at the beginning, being able to position the monad in context with others, in a specific spatial and temporal coordinate, working as universe-building-celular-automata governed by the simplest central rules. 

But using Wikipedia we see some of the real meanings of the word:

* In philosophy: An ultimate atom, or simple, unextended point; something ultimate and indivisible. 

* In functional programming: A data type which represents a specific form of computation, along with the operations "return" and "bind". 


* In category theory: A monoid object in the category of endo-functors of a fixed category. 多多多What???

* Co-monad as  a monad of the opposite category. This makes me think for example "monads" 1 with -1 as the main creators of the category positive and negative integers, but also as a sense of direction in space like spin. 


  For more : https://en.wiktionary.org/wiki/monad


For context we could digress to metaphysics a bit, and remember about Leibiniz "La Monadologie" in the sense that 
what is being proposed  can be seen as a modification of occasionalism developed by latter-day Cartesians. 

* Leibniz surmised that there are indefinitely many substances individually 'programmed' to act in a predetermined way, each substance being coordinated with all the others. 

* This is the pre-established harmony which solved the mind-body problem, but at the cost of declaring any interaction between substances a mere appearance.
