# Intro 





### LaTeX Code


$$
\frac{d}{dt}
\begin{bmatrix}
x(t) \\
y(t) \\
\dot{x}(t) \\
\dot{y}(t)
\end{bmatrix}
=
\begin{bmatrix}
\dot{x}(t) \\
\dot{y}(t) \\
-\frac{x(t)}{(x(t)^2 + y(t)^2)^{3/2}} \\
-\frac{y(t)}{(x(t)^2 + y(t)^2)^{3/2}}
\end{bmatrix}
$$


### Solving the System

This system of differential equations describes the motion of a particle under a central force inversely proportional to the square of the distance from the origin (resembling a Kepler problem or gravitational attraction). The solution involves understanding the dynamics and integrating the equations of motion.

Let's denote the radial distance $ r(t) = \sqrt{x(t)^2 + y(t)^2} $. The acceleration terms $\ddot{x}(t)$ and $\ddot{y}(t)$ are given by:

$$
\ddot{x}(t) = -\frac{x(t)}{(x(t)^2 + y(t)^2)^{3/2}} = -\frac{x(t)}{r(t)^3}
$$

$$
\ddot{y}(t) = -\frac{y(t)}{(x(t)^2 + y(t)^2)^{3/2}} = -\frac{y(t)}{r(t)^3}
$$

These represent the components of the force which is central and attractive, pointing towards the origin. The conserved quantity in such a system is the angular momentum and the energy.

#### Step-by-step Solution:

1. **Angular Momentum Conservation:**

   The angular momentum $ \mathbf{L} = \mathbf{r} \times \mathbf{p} $ (where $\mathbf{p}$ is the linear momentum) is conserved.

   $$
   L = x(t) \dot{y}(t) - y(t) \dot{x}(t)
   $$

   Since $ L $ is constant, let it be $ L_0 $.

2. **Energy Conservation:**

   The total energy (kinetic + potential) is conserved.

   $$
   E = \frac{1}{2} (\dot{x}(t)^2 + \dot{y}(t)^2) - \frac{1}{r(t)}
   $$

   Since $ E $ is constant, let it be $ E_0 $.

3. **Solving for $ r(t) $:**

   Using polar coordinates where $ x(t) = r(t) \cos \theta(t) $ and $ y(t) = r(t) \sin \theta(t) $:

   $$
   \dot{r}(t)^2 + r(t)^2 \dot{\theta}(t)^2 = 2E_0 + \frac{2}{r(t)}
   $$

   With $ \dot{\theta}(t) = \frac{L_0}{r(t)^2} $:

   $$
   \dot{r}(t)^2 + \frac{L_0^2}{r(t)^2} = 2E_0 + \frac{2}{r(t)}
   $$

   This differential equation can be solved to find $ r(t) $, and subsequently $ x(t) $ and $ y(t) $ can be determined from $ r(t) $ and $ \theta(t) $.

Solving these explicitly often involves more advanced techniques and numerical methods for precise trajectories. However, this setup lays out the path to the solution, highlighting the importance of conserved quantities in such central force problems.
